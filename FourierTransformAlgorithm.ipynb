```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
from numpy import pi
import scipy.fftpack as fourier
```

```python

def entradax(t,f):
    #x=(np.cos(2*np.pi*f*t))
    x=(np.sin(2*np.pi*f*t)-0.5 * np.sin(2*np.pi*2*f* t))
    #x=(np.cos(2*np.pi*f*t)-0.5 * np.sin(2*np.pi*2*f* t))
    #x=(np.cos(2*np.pi*f*t)-0.5 * np.sin(2*np.pi*2*f* t))+np.cos(2*10*f*np.pi*t)
    return(x)

n= 2**10
f= 200
tmin=0
tmax=1
t=np.arange(tmin, tmax, ((tmax-tmin)/n))
Ts=t[2]-t[1]
Fs=1/Ts
#x=2*np.exp(-4*t)*np.cos(2*np.pi*f*t)*(t>=0)*1.0
x=entradax(t,f)
X=(np.fft.fft(x)*Ts)
X1=np.fft.fftshift(X)
fqr=np.arange(-Fs/2,Fs/2,Fs/n)
plt.figure()
plt.plot(fqr, abs(X1))
plt.show()
```
